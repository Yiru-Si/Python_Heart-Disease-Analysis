## 2) Data Quality and Consistency Checks

```python
# Check for mixed type
for col in df_heart.columns.tolist():
    weird = (df_heart[[col]].map(type) != df_heart[[col]].iloc[0].apply(type)).any(axis = 1)
    if len (df_heart[weird]) > 0:
        print(col)
```

*No mixed type was observed.*

```python
# Find missing values
df_heart.isnull().sum()
```

*No missing value were found.*

```python
# Find duplicates
df_heart.duplicated()
```
